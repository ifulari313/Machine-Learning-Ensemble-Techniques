```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("titanic_train.csv")
df.head()
```

```python
df2=df[['Survived','Pclass','Age','Parch']]
df2.head()
```

```python
df3=df2.fillna(df2.mean())
df3
```

```python
# now we have to split data set in independat and dependat feature in X and y i.e matrix and vector form
```

```python
X = df3.drop("Survived",axis=1)     # independat
y = df3["Survived"]          # dependant
```

```python
print("shappe of X: ",X.shape)      # 3 independatafeature in X and 1 dependat feature in y
print("shappe of y: ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=51)

## give randomly if not uses random_state , to keep this constant

print("shappe of X_train: ",X_train.shape)
print("shappe of X_test: ",X_test.shape)
print("shappe of y_train: ",y_train.shape)
print("shappe of y_test: ",y_test.shape)
```
